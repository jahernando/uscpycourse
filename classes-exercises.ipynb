A draft of a Vector class

```python
from math import *

class Vector3D:
    """ 3D Vector class 
    """
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
        
    def __str__(self):
        s ='(' + str(self.x) + ', ' + str(self.y) + ', ' + str(self.z) + ')'
        return s
    
    def __add__(self, v):
        x = self.x + v.x
        y = self.y + v.y
        z = self.z + v.z
        return Vector3D(x, y, z)
    
```

```python
# testing the Vector class

def test_vector3d_add(v1, v2):
    v = v1 + v2
    x = v1.x + v2.x
    y = v1.y + v2.y
    z = v1.z + v2.z
    
    ok = (v.x == x) and (v.y == y) and (v.z == z)
    print(str(v)+' == '+str(v1)+' + '+str(v2)+'? '+str(ok))
    return ok
```

```python
vi = Vector3D(1, 0, 0)
vj = Vector3D(0, 1, 0)
vk = Vector3D(0, 0, 1)

test_vector3d_add(vi, vj)
```
